```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

x = np.linspace(-2*np.pi,2*np.pi)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
def func(x,n):
    return x**n

x = np.logspace(-2,2)

plt.plot(x, func(x,1), label='$y=x$')
plt.plot(x, func(x,2), label='$y=x^2$')
plt.plot(x, func(x,3), label='$y=x^3$')

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
N = 1000
np.random.seed(0)
R = np.random.rand(N)

plt.hist(R, bins=100,normed=True)
plt.show()
```

```python
NSTEP = 10000

plt.xlabel('number of steps')
plt.ylabel('position')

for nseed in range(10):
    np.random.rand(nseed)
    step = np.random.choice([-1,1], NSTEP)
    position = np.cumsum(step)
    plt.plot(position)
```
